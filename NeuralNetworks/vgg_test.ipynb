```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from PIL import Image
```

```python
model = keras.applications.VGG19()
```

```python
img = Image.open("images/husky_224.jpg")
```

```python
plt.imshow(img)
```

```python
img_array = np.array(img)
x = keras.applications.vgg19.preprocess_input(img_array)
x = np.expand_dims(x, axis=0)
print(x.shape)
```

```python
res = model.predict(x)
np.argmax(res)
```

```python
labels_path = tf.keras.utils.get_file(
'imagenet1000_clsidx_to_labels.txt','https://gist.githubusercontent.com/yrevar/942d3a0ac09ec9e5eb3a/raw/238f720ff059c1f82f368259d1ca4ffa5dd8f9f5/imagenet1000_clsidx_to_labels.txt')
imagenet_labels = np.array(open(labels_path).read().splitlines())
predicted_class_name = imagenet_labels[np.argmax(res)]
predicted_class_name
```
